```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
#Import Modules

# Memasukkan dataset
dataset_2816 = pd.read_csv('Dataset_2816.csv')
YearsExperience_2816 = dataset_2816.iloc[:, :-1].values #Mendapatkan Copy dataset kecuali kolom terakhir
Salary_2816 = dataset_2816.iloc[:, 1].values #Mendapatkan array dataset dalam kolom pertama
dataset_2816 #Memunculkan dataset

X_train = YearsExperience_2816
y_train = Salary_2816

regressor_2816 = LinearRegression()
regressor_2816.fit(X_train, y_train)

Reihan_2816 = plt
Reihan_2816.scatter(X_train, y_train, color='red')
Reihan_2816.plot(X_train, regressor_2816.predict(X_train), color='black')
Reihan_2816.title('Salary VS Experience (Prediction)')
Reihan_2816.xlabel('Year of Experience')
Reihan_2816.ylabel('Salary')
Reihan_2816.show()

# Memprediksi hasil test dengan pengalaman kerja 5 tahun
y_pred = regressor_2816.predict(np.array([5]).reshape(1, 1))
print("You can offer to your candidate the salary of $", y_pred, "and this is the best salary for him!")
```
